```python
from bs4 import BeautifulSoup
import logging
import numpy as np
import pandas as pd
import requests
from tqdm import tqdm
```

```python
link_template = 'https://sigir2025.dei.unipd.it/accepted-papers.html'
def get_title_author(link):
    resp = requests.get(link, proxies=None)
    soup = BeautifulSoup(resp.content, 'html.parser')
    uls = soup.find_all('ul')
    titles, authors = [], []
    for ul in uls:  # 遍历所有 ul
        articles = ul.find_all('li', class_='accepted-paper-item')  # 找到当前 ul 下的所有 li
        print(f"Found {len(articles)} articles in one ul.")
        for li in articles:
            title = li.find('span', class_='accepted-paper-title').text
            author = li.find('span', class_='accepted-paper-author').text
            titles.append(title)
            authors.append(author)
    return titles, authors
```

```python
titles, authors = get_title_author(link_template)
print(titles, authors)
```

```python
print(len(titles), len(authors))
```

```python
def graph_mask(title):
    t = title.lower()
    conditions = ['graph neural networks', 'gnns', 'gnn', 'graph', 'graph learning', 'graph embedding', 'network embedding']
    return np.any([c in t for c in conditions])
```

```python
with open(f'2025-SIGIR.txt', 'w', encoding='utf8') as fw:
    for t, n in zip(titles, authors):
        if graph_mask(t):
            fw.write(f'1. **{t}**\n\n')
            fw.write(f'    *{n}*\n\n')
```
