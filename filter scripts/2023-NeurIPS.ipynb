```python
from bs4 import BeautifulSoup
import logging
import numpy as np
import pandas as pd
import requests
from tqdm import tqdm
```

```python
link_template = 'https://proceedings.neurips.cc/paper/2024'
def get_title_author(link):
    resp = requests.get(link, proxies=None)
    soup = BeautifulSoup(resp.content, 'html.parser')
    articles = soup.find('ul', class_='paper-list')
    titles, authors = [], []
    articles = articles.find_all('li')
    print(len(articles))
    for art in tqdm(articles):
        title = art.find('a').text.strip()
        author = art.find('i').text.strip()
        titles.append(title)
        authors.append(author)
    return titles, authors
```

```python
titles, authors = get_title_author(link_template)
print(titles, authors)
```

```python
print(len(titles), len(authors))
```

```python
def graph_mask(title):
    t = title.lower()
    conditions = ['graph neural networks', 'gnns', 'gnn', 'graph', 'graph learning', 'graph embedding', 'network embedding']
    return np.any([c in t for c in conditions])
```

```python
with open(f'2024-NeurIPS.txt', 'w', encoding='utf8') as fw:
    for t, n in zip(titles, authors):
        if graph_mask(t):
            fw.write(f'1. **{t}**\n\n')
            fw.write(f'    *{n}*\n\n')
```
